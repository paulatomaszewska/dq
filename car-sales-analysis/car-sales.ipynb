## Car sales
This is a simple project in which I'll clean data and do some basic analysis.

```python
import pandas as pd
import numpy as np
```

```python
autos = pd.read_csv("autos.csv", encoding="latin1")
autos.info()
autos.head()
```

### Cleaning column names

```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'odometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'nr_of_pictures', 'postal_code',
       'last_seen']

autos.head()
```

### Exploring the dataset

```python
autos.describe(include="all")
```

#### Dropping columns that have mostly one value, so won't be useful for analysis

```python
autos = autos.drop(labels=["seller", "offer_type", "nr_of_pictures"], axis=1)
autos.head()
```

#### Dealing with the numerical columns that are represented as strings

```python
autos["price"] = autos["price"] \
    .str.replace("$", "") \
    .str.replace(",","") \
    .astype(int)

autos["odometer"] = autos["odometer"] \
    .str.replace("km", "") \
    .str.replace(",","") \
    .astype(int)

autos = autos.rename(mapper={"price": "price_usd", "odometer": "odometer_km"}, axis=1)
autos.head()
```

#### Removing weird values of price (cheaper than 100$, more expensive than 10 mln $)

```python
autos["price_usd"].value_counts().sort_index(ascending=False).head(20)

autos = autos.loc[autos["price_usd"].between(100,10000000)]
```

#### Exploring dates

1) `date_crawled` - we can see that the crawler ran between 05.03.2016 and 07.04.2016
2) `ad_created` - the ads were created between 11.06.2015 and 07.04.2016
3) `last_seen` - those dates are the same as `date_crawled` - which makes sense, because this date also comes from the crawler.
4) `registration_year` - we can see that there are some impossible values, like 1000 or 9999. I will clean values occurring before 1920 and after 2016 when the crawler ran. I chose 1920 because this is the earliest I can see an actual vintage car in this dataset, the rest seems like they are just random numbers.
5) `registration_month` - there is a strange value (month "0"). I decided to keep it and treat it as "unknown". The cars with this value look valid, we just don't seem to know the correct registration month.


```python
autos = autos.loc[autos["registration_year"].between(1920, 2016)]
```

#### Exploring `price` by `brand`
Now I'd like to explore the price by brand. For this, I'm choosing the brands which make for over 5% of all ads.

We can see that on average, Audis are the most expensive ones, followed by Mercedes and BMW. The cheapers of the analyzed brands is Opel. The average mileages are similar between the brands.

```python
price_averages_by_brand = {}
mileage_averages_by_brand = {}

for brand in autos["brand"].value_counts(normalize=True)[:6].index:
    prices_per_brand = autos.loc[autos["brand"] == brand, "price_usd"]
    price_averages_by_brand[brand] = round(prices_per_brand.sum() / prices_per_brand.count())

    mileages_per_brand = autos.loc[autos["brand"] == brand, "odometer_km"]
    mileage_averages_by_brand[brand] = round(mileages_per_brand.sum() / mileages_per_brand.count())



price_averages_by_brand_series = pd.Series(price_averages_by_brand, name="avg_price")
mileage_averages_by_brand_series = pd.Series(mileage_averages_by_brand, name="avg_mileage")

pd.concat([price_averages_by_brand_series, mileage_averages_by_brand_series], axis=1)
```
